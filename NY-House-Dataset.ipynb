```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("NY-House-Dataset.csv")
```

```python
df
```

```python
df.nunique()
```

```python
df.dropna(inplace=True)
df
```

```python
df = pd.get_dummies(df,dtype='int')
df
```

```python
X = df.drop(columns=['PRICE'])
Y = df['PRICE']
X
```

```python
Y
```

```python

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
model_lin = LinearRegression()
model_lin.fit(X_train,Y_train)
```

```python
model_lin.score(X_train,Y_train)
```

```python
model_lin.score(X_test,Y_test)
```

```python
model_las = Lasso(alpha = 50,max_iter = 100,tol = 0.1)
```

```python
model_las.fit(X_train,Y_train)
```

```python
model_las.score(X_train,Y_train)
```

```python
model_las.score(X_test,Y_test)
```

```python
model_rid  = Ridge(alpha = 1,max_iter = 1,tol = 0.001)
```

```python
model_rid.fit(X_train,Y_train)
```

```python
model_rid.score(X_train,Y_train)
```

```python
model_rid.score(X_test,Y_test)
```
